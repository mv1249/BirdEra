```python
import os
import random
import shutil
```

```python
train_len = os.listdir('train')
len(train_len)
```

```python
os.mkdir('Training')
os.mkdir('Testing')
```

```python
from glob import glob
```

```python
train_mat = [i for i in glob('train/*')]
test_mat = [i for i in glob('test/*')]
train_mat
```

```python
train_list = [i for i in os.listdir('train')]
test_list = [i for i in os.listdir('test')]
train_list
```

```python
training = []
testing = []
train = 'Training\\'
for i in train_list:
    training.append(train+i)
```

```python
test_new = []
test = 'Testing\\'
for i in test_list:
    test_new.append(test+i)
```

```python
for i in training:
    os.mkdir(i)
for i in test_new:
    os.mkdir(i)
```

```python
train_mat
```

```python
test_mat
```

```python
for i in range(150):
    for j in range(50):
        img_path = train_mat[i]
        to_path = training[i]
        images = os.listdir(img_path)
        random.shuffle(images)
        
        image_name = images[j]
        img_new_path = os.path.join(img_path,image_name)
        dest_path = os.path.join(to_path,image_name)
        shutil.copy2(img_new_path,dest_path)
        
```

```python
for i in range(150):
    for j in range(3):
        img_path = test_mat[i]
        to_path = test_new[i]
        images = os.listdir(img_path)
        random.shuffle(images)
        
        image_name = images[j]
        img_new_path = os.path.join(img_path,image_name)
        dest_path = os.path.join(to_path,image_name)
        shutil.copy2(img_new_path,dest_path)
        
```
